Copy and paste the code from the Week 9 moving window notebook that generates the TTR CSV files using the moving window approach. You will also need to modify the code so that instead of computing 4 different moving windows, it only computes one with a window size of 500. This means that you should remove the inner for loop so that `moving_TTR()` is only called once.

The column heading for the TTR will be `"MW TTR 560"`

The moving window TTR function is given below.  You don't need to modify it.

After running the code, you should see a CSV file named `ttr-windows.csv` in the same folder as this notebook.

**Note**: Do not use the code in the Week 5 lecture notebook — be sure to use the code from the **Week 9** moving window lecture notebook.

```python
from collections import Counter

def moving_TTR(tokens, window_size):
    n = len(tokens)
    if window_size <= 0 or window_size > n:
        return 0

    # initialize first window
    window = tokens[:window_size]
    counts = Counter(window)            # word -> frequency in current window
    # counts is a dictionary that maps each distint work in window to the number of times that word appears

    num_types = len(counts)              # number of types in current window
    types_sum = num_types

    # slide the window once per step
    for i in range(window_size, n):
        out_tok = tokens[i - window_size]
        in_tok  = tokens[i]

        # remove leftmost token
        counts[out_tok] -= 1
        if counts[out_tok] == 0:
            del counts[out_tok]
            num_types -= 1

        # add rightmost token
        prev = counts.get(in_tok, 0)
        counts[in_tok] = prev + 1
        if prev == 0:
            num_types += 1

        types_sum += num_types

    num_windows = n - window_size + 1
    return types_sum / (num_windows * window_size) * 100
```
